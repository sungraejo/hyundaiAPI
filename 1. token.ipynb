```python
client_id = "SET_YOUR_CLIENT_ID"
client_secret = "SET_YOUR_CLIENT_SECRET"

hyundai_id = "SET_YOUR_HYUNDAI_ID"
hyundai_pw = "SET_YOUR_HYUNDAI_PW"

redirect_uri = "https://cpp13.blogspot.com/"
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

auth_url = f"https://prd.kr-ccapi.hyundai.com/api/v1/user/oauth2/authorize?response_type=code&client_id={client_id}&redirect_uri={redirect_uri}&state=state"
print(auth_url)

driver = webdriver.Chrome()
driver.get(auth_url)
driver.implicitly_wait(3)

inputElement = driver.find_element(By.XPATH, "//*[@id='root']/div/div[2]/div[1]/div/div/div[1]/div[1]/div/div/input")
inputElement.send_keys(hyundai_id)
inputElement = driver.find_element(By.XPATH, "//*[@id='root']/div/div[2]/div[1]/div/div/div[1]/div[2]/div/div/input")
inputElement.send_keys(hyundai_pw)
inputElement = driver.find_element(By.XPATH, "//*[@id='root']/div/div[2]/div[1]/div/div/div[1]/div[2]/div/div/input")
buttonElement = driver.find_element(By.XPATH, "//*[@id='root']/div/div[2]/div[1]/div/div/div[2]/button")
buttonElement.click()
driver.implicitly_wait(3)

carButtonElement = driver.find_element(By.XPATH, "//*[@id='root']/div/div[3]/div/div/div/ul/div")
carButtonElement.click()
agreeuttonElement = driver.find_element(By.XPATH, "//*[@id='root']/div/div[5]/div/div/button")
agreeuttonElement.click()

oauth_url = driver.current_url
print(oauth_url)
driver.quit()
```

```python
from urllib import parse
from urllib.parse import urlparse, parse_qs

url = parse.urlparse(oauth_url)
code = parse_qs(url.query)['code'][0]
print(code)
```

```python
import json
import requests
import base64

auth_str = f"{client_id}:{client_secret}"
auth_bytes = auth_str.encode("utf-8")
auth_base64 = base64.b64encode(auth_bytes).decode("utf-8")

session = requests.Session()
resp = session.post(
    url="https://prd.kr-ccapi.hyundai.com/api/v1/user/oauth2/token",
    headers={
        'authorization': "Basic %s" % auth_base64
    },
    data={
        'grant_type': 'authorization_code',
        'code': code,
        'redirect_uri': redirect_uri
    }
)

print(resp.status_code)
print(json.dumps(resp.json(), ensure_ascii=False, indent=3))
```

```python
access_token = resp.json()["access_token"]
print(access_token)
```

```python
f = open("./token", 'w')
f.write(access_token)
f.close()
```
