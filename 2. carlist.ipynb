```python
import requests
import json
import urllib.parse
session = requests.Session()
```

```python
f = open("./token", 'r')
auth = f.readline()
f.close()
```

```python
resp = session.get(
    url="https://dev.kr-ccapi.hyundai.com/api/v1/car/profile/carlist",
    headers={
        'authorization': "Bearer %s" % auth
    }
)
print(resp.status_code)
print(json.dumps(resp.json(), ensure_ascii=False, indent=3))
```

```python
cars = resp.json()["cars"]
carId = cars[0]['carId']
print(carId)

f = open("./carId", 'w')
f.write(carId)
f.close()
```
