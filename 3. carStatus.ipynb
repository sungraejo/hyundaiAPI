```python
import requests
import json
import urllib.parse
session = requests.Session()
```

```python
f = open("./token", 'r')
auth = f.readline()
f.close()

f = open("./carId", 'r')
carId = f.readline()
f.close()
```

```python
resp = session.get(
    url="https://dev.kr-ccapi.hyundai.com/api/v1/car/status/%s/dte" % carId,
    headers={
        'authorization': "Bearer %s" % auth
    }
)

print(resp.status_code)
print(json.dumps(resp.json(), ensure_ascii=False, indent=3))
```
